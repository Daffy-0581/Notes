## DATA MANIPULATION 

```python
# isnull -> returns missing value as True and False
dataset.isnull()
```

```python
# reverse of isnull()
dataset.notnull()
```

```python
# count of missing rows
# 0 - no missing values in row
# 1 - missing values in row

dataset.isnull().sum()
```

```python
# sum of missing rows present in each column

dataset.isna().sum()
```

```python
# Describe by default applies to numerical columns

dataset.describe()
```

```python
# replace '?' by np.nan
# inplace = True - (Apply changes in '?' by np.nan)
dataset.replace("?", np.nan, inplace=True)
dataset.isnull().sum()
```

```python
# select 'price' column
dataset['price']
```

```python
# selecting multiple columns
dataset[['make','peak-rpm','price','wheel-base']]
```

```python
# returns true where value is null else false
dataset['price'].isnull()
```

```python
dataset[dataset['price'].isnull() == True]
```

```python
# data location - does not read value as index hence displays all values includig 10
dataset.loc[0:10,]
```

```python
# start index = 'make' , stop index = 'height'
dataset.loc[0:10,'make':'body-style']
```

```python
dataset.loc[0:10,['make','fuel-type','num-of-doors','body-style']]
```

```python
# iloc - takes input as indexes and not the column attributes
dataset.iloc[2:15,2:13]
```

### drop()   -  dropping rows, dropping columns

```python
# axis = 0 : drop row
dataset.drop([0,1,2,3,4,5,6,7,8], axis=0)
```

```python
# axis = 1 : drop column
dataset.drop(['symboling','normalized-losses'], axis=1, inplace=True)
```

```python
dataset.head()
```

```python
dataset.drop([0,1,2,3,4,5,6,7,8], axis=0).reset_index().drop('index', axis = 1)
```

### filter()

```python
dataset[dataset['make'] == "audi"]
```

```python
# filter by one condition
dataset.filter(items = ["make","audi"])
```

```python
# filter by multiple conditions using & operator
dataset[(dataset['make'] == "audi") & (dataset['horsepower'] > 90)]
```

```python
# filter by multiple conditions using | operator
dataset[(dataset['make'] == "audi") | (dataset['horsepower'] > 90)]
```

### duplicate()

### Create a 3 column (col1, col2, col3) dataframe using pandas series & pd.DataFrame() - atleast 2 rows should have same values
### Apply duplicated() , dataset[dataset.duplicated()], dataset[dataset.duplicated()].index

```python
data_series = {'Column 1' : pd.Series([10,20,30,20,10,60,70]),
               'Column 2' : pd.Series([100,200,300,200,100,600,700]),
               'Column 3' : pd.Series([100,200,300,200,100,600,700])}

df = pd.DataFrame(data_series)
df
```

```python
df.duplicated()
```

```python
df[df.duplicated()]
```

```python
df[df.duplicated()].index
```
